```python
import seaborn as sns
iris=sns.load_dataset("iris")
iris.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
#ANS-
#1):-scatter plot:-A scatter plot is used to visualize the relationship between two variables. It is useful in determining whether the variables are correlated or not. 
                   #Seaborn's scatter plot allows us to add a third variable to the plot using the hue parameter.
print(sns.scatterplot(x=iris.petal_length,y=iris.sepal_length))
plt.title("scatter plot")


#2):-displot:-It is used basically for univariant set of observations and visualizes it through a histogram 
             #i.e. only one observation and hence we choose one particular column of the dataset.
print(sns.displot(iris["sepal_length"]))
plt.title("displot")


#3):-relplot:-This function provides us the access to some other different axes-level functions which shows the relationships between two variables with semantic mappings of subsets.
print(sns.relplot(x=iris.petal_length,y=iris.sepal_length,data=iris,hue="species"))
plt.title("relplot")


#4):-catplot:-This function provides access to several axes-level functions that show the relationship between a numerical and
              #one or more categorical variables using one of several visual representations. 
print(sns.catplot(x=iris.species,y=iris.sepal_length,data=iris))
plt.title("catplot")


#5):-pairplot:-To plot multiple pairwise bivariate distributions in a dataset, you can use the .pairplot() function. 

#The diagonal plots are the univariate plots, and this displays the relationship for the (n, 2) combination of variables in a DataFrame as a matrix of plots.
print(sns.pairplot(iris))
plt.title("pairplot")
```

```python
#Que 2: Load the "fmri" dataset using the load_dataset function of seaborn.
        #Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

#ANS-
fmri=sns.load_dataset("fmri")
sns.lineplot(x=fmri.timepoint,y=fmri.signal)
```

```python
#Que 3: Load the "titanic" dataset using the load_dataset function of seaborn.
        #Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

#ANS-
titan=sns.load_dataset("titanic")
sns.boxplot(x=titan.pclass,y=titan.age)
sns.boxplot(x=titan.pclass,y=titan.fare)
```

```python
#Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. 
        #Use the hue parameter for the 'cut' column of the diamonds dataset.
#ANS-
dia=sns.load_dataset("diamonds")
sns.displot(x=dia.price,hue ="cut",data=dia)
```
